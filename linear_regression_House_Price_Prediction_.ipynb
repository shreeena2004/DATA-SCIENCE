```python
import warnings
warnings.filterwarnings("ignore")

from scipy.stats import norm
sns.distplot(house_data['price'],fit = norm )
```

```python
sns.boxplot(house_data['price'],orient='h')
```

```python
house_df = house_data.drop(['id','date'],axis = 1)
```

```python
!pip install zscore
```

```python
#Eliminate outliers
from scipy.stats import zscore
z = zscore(house_df)
```

```python
np.where(z > 3)[0]
```

```python
np.where(z > 4)
```

```python
house_df1 = house_df.drop(np.where(z > 4)[0])
```

```python
sns.distplot(house_df1['price'], fit = norm)
```
