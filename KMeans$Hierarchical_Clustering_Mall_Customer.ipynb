##### feature


```python
X = mall_customers.iloc[:,3:5]
```

```python
X
```

##### Elbow Method
plot between kwss and k (range of clusters)

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,16):
  kmeans = KMeans(n_clusters = i, init = 'k-means++', n_init = 10)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)  #wcss is returned by inertia
```

```python
plt.plot(range(1,16),wcss)
plt.scatter(x = range(1,16), y = wcss, color = 'black',marker = 'o')
plt.show()
```

#### FinalCluster using k = 5

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++',n_init =10,max_iter = 300)
kmeans.fit(X)
```

```python
# clusster centorids
kmeans.cluster_centers_
```

```python
y_predict = kmeans.predict(X)
```

```python
X['Clusters'] = y_predict
```

```python
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Clusters', data=X, palette='Set1')
sns.scatterplot(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:, 1], c='black', marker='x', s=100)
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
fig = plt.figure(figsize= (15,5))
dendrogram = sch.dendrogram(sch.linkage(X,method = 'ward'))
plt.axhline(y = 120, linestyle = '--',color ='k')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5 , linkage = 'ward', affinity = 'euclidean')
```

```python
hc.fit(X)
```

```python
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

```python
X['Clusters'] = y_hc
```

```python
plt.figure(figsize = (12,8))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Clusters', data=X, palette='Set1')
plt.show()
```
